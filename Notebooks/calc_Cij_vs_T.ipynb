###1.3 Calculation Parameters

These are user-specified parameters that control how the calculation proceeds, such as number of data points.

- run_steps is the total number of MD iteration steps to run at each temperature. Increasing the run time adds to the number of data points being averaged and can help improve the statistical analysis.

- relax_steps is the number of MD iteration steps at the beginning of the run to ignore.  This allows for the data during the initial relaxation period to be excluded. 

- dump_every sets how often LAMMPS dump files of the atoms and calculated parameters are created.  By default, setting dump_every = run_steps means only two files are created: one for the initial configuration and one for the final. This helps in the analysis associated with determining if the crystal structure has transformed. 

- thermo_steps sets how often the system-wide measurements are made.  Decreasing this number adds to the number of data points being averaged, but doesn't add to the total run time.

- size sets the dimensions of the system that is created.  Increasing the system size decreases the deviation noise for the values measured, but also greatly increases the calculation time.  

- random_seed is an integer that is used by LAMMPS in generating random numbers for assigning velocities.  If set to None, a random number is assigned to the random_seed.

##3. LAMMPS Script Generation Function(s)

This code generates the underlying LAMMPS script(s) for performing the necessary simulations.  

- temp_cij_script() performs NVT simulations of the unstrained system, and +- small strains for each of the 6 strain directions (xx, yy, zz, yz, xz, xy). 

```python
def temp_cij_script(pair_info, units, atom_style, masses, ucell, alat, 
                    delta = 1e-3, steps = 2, size = np.array([3, 3, 3], dtype=np.int),
                    temperature = 0, run_steps = 100000, dump_every = 100000, thermo_steps = 1000,
                    random_seed = None):
    nl = '\n'
    
    boxsize = np.array([[0,size[0]], [0,size[1]], [0,size[2]]], dtype=np.int)
    sys_gen = iprp.lammps.sys_gen_script(alat=alat, ucell=ucell, size=boxsize)
    
    mass = ''
    for i in xrange(0,len(masses)):
        mass += 'mass %i %f\n'%(i+1, masses[i])
    
    if random_seed == None:
        random_seed = random.randrange(1,1000000)
    
    if temperature == 0.0:
        integrate = nl.join(['fix nve all nve',
                             'fix langevin all langevin 0.0 0.0 1.0 %d' % (random_seed)])
    else:
        integrate = nl.join(['velocity all create %lf %d' % (2*temperature, random_seed),
                             'fix nvt all nvt temp %lf %lf 1.0' % (temperature, temperature)])
    
    script = nl.join(['boundary p p p',
                      'units '+ units,
                      'atom_style ' + atom_style,
                      '',
                      sys_gen,
                      '',
                      mass,
                      pair_info,
                      '',
                      'variable lx0 equal $(lx)',
                      'variable ly0 equal $(ly)',
                      'variable lz0 equal $(lz)',
                      '',
                      'variable deltax equal %f/%f' % (delta, steps-1),
                      '',
                      'variable alat equal lx/%i' % (size[0]),
                      'variable blat equal ly/%i' % (size[1]),
                      'variable clat equal lz/%i' % (size[2]),
                      '',
                      'variable peatom equal pe/atoms',
                      '',
                      'thermo %i' % (thermo_steps),
                      'thermo_style custom step lx ly lz yz xz xy pxx pyy pzz pyz pxz pxy etotal',
                      'thermo_modify format float %.13e',
                      '',
                      integrate,
                      '',
                      'run %i' % run_steps,
                      '',
                      'variable aratio equal 1-%f/2.+(v_a-1)*${deltax}' % (delta),
                      'variable xmax equal v_aratio*${lx0}',
                      'label loopa',
                      'variable a loop %i' % (steps),
                      'change_box all x final 0 ${xmax} remap units box',
                      'run %i' % run_steps,
                      'next a','jump temp_cij.in loopa',
                      'change_box all x final 0 ${lx0} remap units box',
                      '',
                      'variable bratio equal 1-%f/2.+(v_b-1)*${deltax}' % (delta),
                      'variable ymax equal v_bratio*${ly0}',
                      'label loopb',
                      'variable b loop %i' % (steps),
                      'change_box all y final 0 ${ymax} remap units box',
                      'run %i' % run_steps,
                      'next b','jump temp_cij.in loopb',
                      'change_box all y final 0 ${ly0} remap units box',
                      '',
                      'variable cratio equal 1-%f/2.+(v_c-1)*${deltax}'%(delta),
                      'variable zmax equal v_cratio*${lz0}',
                      'label loopc',
                      'variable c loop %i' % (steps),
                      'change_box all z final 0 ${zmax} remap units box',
                      'run %i' % run_steps,
                      'next c','jump temp_cij.in loopc',
                      'change_box all z final 0 ${lz0} remap units box',
                      '',
                      'change_box all triclinic',
                      'variable eyz equal (-%f/2.+(v_d-1)*${deltax})*${lz0}' % (delta),
                      'label loopd',
                      'variable d loop %i' % (steps),
                      'change_box all yz final ${eyz} remap units box',
                      'run %i' % run_steps,
                      'next d','jump temp_cij.in loopd',
                      'change_box all yz final 0 remap units box',
                      '',
                      'variable exz equal (-%f/2.+(v_e-1)*${deltax})*${lz0}'%(delta),
                      'label loope',
                      'variable e loop %i' % (steps),
                      'change_box all xz final ${exz} remap units box',
                      'run %i' % run_steps,
                      'next e','jump temp_cij.in loope',
                      'change_box all xz final 0 remap units box',
                      '',
                      'variable exy equal (-%f/2.+(v_f-1)*${deltax})*${ly0}' % (delta),
                      'label loopf',
                      'variable f loop %i '% (steps),
                      'change_box all xy final ${exy} remap units box',
                      'run %i' % run_steps,
                      'next f','jump temp_cij.in loopf',
                      'change_box all xy final 0 remap units box'])

    return script
```

##4. Python Calculation Function(s)

These functions are associated with setting up the atomic systems, running LAMMPS, extracting the simulation data, and refining the results.

- print_cij() prints out the elastic constant matrix in a pretty and readable format.

```python
def print_cij(o_cij):
    cij = np.empty((6,6))
    for i in xrange(6):
        for j in xrange(6):
            if np.isclose(o_cij[i,j], 0.0, rtol=0, atol=0.01):
                cij[i,j] = 0.0
            else:
                cij[i,j] = o_cij[i,j]
    
    print '[[%7.2f, %7.2f, %7.2f, %7.2f, %7.2f, %7.2f],' % (cij[0,0], cij[0,1], cij[0,2], cij[0,3], cij[0,4], cij[0,5])
    print ' [%7.2f, %7.2f, %7.2f, %7.2f, %7.2f, %7.2f],' % (cij[1,0], cij[1,1], cij[1,2], cij[1,3], cij[1,4], cij[1,5])
    print ' [%7.2f, %7.2f, %7.2f, %7.2f, %7.2f, %7.2f],' % (cij[2,0], cij[2,1], cij[2,2], cij[2,3], cij[2,4], cij[2,5])
    print ' [%7.2f, %7.2f, %7.2f, %7.2f, %7.2f, %7.2f],' % (cij[3,0], cij[3,1], cij[3,2], cij[3,3], cij[3,4], cij[3,5])
    print ' [%7.2f, %7.2f, %7.2f, %7.2f, %7.2f, %7.2f],' % (cij[4,0], cij[4,1], cij[4,2], cij[4,3], cij[4,4], cij[4,5])
    print ' [%7.2f, %7.2f, %7.2f, %7.2f, %7.2f, %7.2f]]' % (cij[5,0], cij[5,1], cij[5,2], cij[5,3], cij[5,4], cij[5,5])
    
```

```python
#Read potential information
pot = iprp.lammps.Potential(pot_file, pot_dir)

#Read crystal prototype information
proto = iprp.models.CrystalPrototype(proto_file)

#Build masses list 
if proto.get('nsites') == len(elements):
    masses = []
    for el in elements:
        masses.append(iprp.get_mass(el))
else:
    raise ValueError('Number of elements listed does not match number of prototype sites')

alat = np.array(alat)

#Get parameters from pot
units = pot.get('units')
pair_info = pot.coeff(elements)
atom_style = pot.get('atom_style') 

#Get parameters from proto
ucell = proto.get('ucell')

#Perform the simulation
with open('temp_cij.in','w') as f:
    f.write(temp_cij_script(pair_info, units, atom_style, masses, ucell, alat, 
                             size=size, temperature=temperature, thermo_steps=thermo_steps,
                             run_steps=run_steps, dump_every=dump_every))

data = iprp.lammps.extract(subprocess.check_output(lammps_exe+' < temp_cij.in',shell=True))

step = np.empty(len(data)-1, dtype=np.int)
l_list = np.empty((len(data)-1, 6))
p_list = np.empty((len(data)-1, 6))

for i in xrange(1, len(data)):
    step[i-1] = int(data[i][0])
    
    for j in xrange(6):
        l_list[i-1][j] = float(data[i][j+1])
        p_list[i-1][j] = -float(data[i][j+7]) * 1e-4

dimensions = np.empty((13,6))
stresses = np.empty((13,6))
stress_errs = np.empty((13,6))
        
end_index = 0
for p in xrange(13):
    start_index = end_index + relax_steps / thermo_steps
    end_index += run_steps / thermo_steps + 1

    for i in xrange(6):
        dimensions[p, i] = l_list[start_index, i]
        stresses[p, i] = np.mean(p_list[start_index : end_index, i])
        stress_errs[p, i] = 2*np.std(p_list[start_index : end_index, i])

print 'Check initial stress state:'
print 'Direction Width(A)   Stress(GPa) P-Error'
print 'XX       %10f %10f %10f' % (dimensions[0,0], stresses[0,0], stress_errs[0,0])
print 'YY       %10f %10f %10f' % (dimensions[0,1], stresses[0,1], stress_errs[0,1])
print 'ZZ       %10f %10f %10f' % (dimensions[0,2], stresses[0,2], stress_errs[0,2])
print 'YZ       %10f %10f %10f' % (dimensions[0,3], stresses[0,3], stress_errs[0,3])
print 'XZ       %10f %10f %10f' % (dimensions[0,4], stresses[0,4], stress_errs[0,4])
print 'XY       %10f %10f %10f' % (dimensions[0,5], stresses[0,5], stress_errs[0,5])
print 
eps_scale = np.array([dimensions[0,0], dimensions[0,1], dimensions[0,2], 
                      dimensions[0,2], dimensions[0,2], dimensions[0,1]])

#Calculate Cij from LAMMPS results
cij = np.empty((6,6))
cij_errs = np.empty((6,6))
for i in xrange(0,6):
    eps_ij = (dimensions[2*i+2, i] - dimensions[2*i+1, i]) / eps_scale[i]  
    for j in xrange(0,6):
        sig_ij = stresses[2*i+2, j] - stresses[2*i+1, j]
        sig_err = ((stress_errs[2*i+2, j])**2 + (stress_errs[2*i+1, j])**2 )**0.5
        cij[j,i] = sig_ij/eps_ij
        cij_errs[j,i] = abs(sig_err/eps_ij)
print 'Cij matrix (GPa) ='
print_cij(cij)    
print
print 'Cij errors (GPa) ='
print_cij(cij_errs)
```
