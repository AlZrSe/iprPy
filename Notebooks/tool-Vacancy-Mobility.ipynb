##3. LAMMPS Script Generation Function(s)

This code generates the underlying LAMMPS script(s) for performing the necessary simulations.  

- neb_script() performs an NEB energy barrier calculation on a system. 

```python
def neb_script(pair_info, units, atom_style, masses, read_data, read_neb, nreplicas,
               spring_const=5.0, etol = 0.0, ftol = 1e-4, maxiter = 1000, maxeval = 1000):
    mass = ''
    for i in xrange(len(masses)):
        mass += 'mass %i %f\n'%(i+1,masses[i])
    
    nl = '\n'
    script = nl.join(['boundary p p p',
                      'units ' + units,
                      'atom_style ' + atom_style,
                      'atom_modify map array sort 0 0.0',
                      'read_data ' + read_data,
                      '',
                      mass,
                      pair_info,
                      '',
                      'variable peatom equal pe/atoms',
                      '',
                      'thermo_style custom step lx pxx pe v_peatom',
                      'thermo_modify format float %.13e',
                      '',
                      'fix NEB all neb %f' % spring_const,
                      'variable u uloop %i' % nreplicas,
                      '',
                      'dump dumpit all custom 100000 atom_$u.* id type x y z',
                      'dump_modify dumpit format "%i %i %.13e %.13e %.13e"',
                      '',
                      'min_style quickmin',
                      'timestep 0.01',
                      '',
                      'neb %f %f %i %i 50 final %s' % (etol, ftol, maxiter, maxeval, read_neb)])
    return script
```

##4. Python Calculation Function(s)

These functions are associated with setting up the atomic systems, running LAMMPS, extracting the simulation data, and refining the results.

- neb_vacancy() sets up, runs and extracts the data from the neb_script run plot the energy barrier.  

```python
#Compute the energies associated with point defects for different system sizes   
def neb_vacancy(sys1, sys2, lammps_exe, pot, elements, masses, nreplicas, 
                spring_const=5.0, etol = 0.0, ftol = 1e-4, maxiter = 1000, maxeval = 1000):
    
    pair_info = pot.coeff(elements)
    units = pot.get('units')
    atom_style = pot.get('atom_style')
    
    #Initialization
    rcoord = np.zeros(nreplicas)
    energy = np.zeros(nreplicas)
    
    #Create start and end state data files
    iprp.lammps.write_atom('start.dat',sys1)
    with open('end.dats','w') as neb_file:
        neb_file.write('%i\n' % len(sys2.atoms))
        for i in xrange(len(sys2.atoms)):
            neb_file.write('%i %.13e %.13e %.13e\n' % (i+1, sys2.atoms[i].get('x'), sys2.atoms[i].get('y'), sys2.atoms[i].get('z')))
    
    #Create LAMMPS input script
    with open('neb.in','w') as f:
        f.write(neb_script(pair_info, units, atom_style, masses, 'start.dat', 'end.dats', nreplicas, 
                           spring_const=spring_const, etol=etol, ftol=ftol, maxiter=maxiter, maxeval=maxeval))
    
    #Run lammps
    mpilammps = 'mpiexec -localonly %i %s -p %ix1 -in neb.in'%(nreplicas, lammps_exe, nreplicas)
    output = subprocess.check_output(mpilammps, shell=True)
    
    
    lines = output.split('\n')
    for line in lines:
        terms = line.split()
        if len(terms)>0:
            lastline = terms
    
    for i in xrange(nreplicas):
        rcoord[i] = float(lastline[9+2*i])
        f = open('log.lammps.'+str(i))
        eread = False
        for line in f:
            terms = line.split()
            if len(terms)>0:
                if eread:
                    energy[i] = float(terms[2])
                    eread = False
                else:
                    if terms[0]=='Energy':
                        eread = True
        f.close()
    energy0 = energy[0]
    for i in xrange(nreplicas):
        energy[i] = energy[i]-energy0
                    
    return rcoord, energy
```

##5. Run Calculation and Display Results

This is the code that runs the calculation function(s) for the supplied inputs and displays the resulting data.

```python
#Read the data model files
pot = iprp.lammps.Potential(pot_file, pot_dir)
proto = iprp.models.CrystalPrototype(proto_file)
ptd = iprp.models.PointDefect(ptd_file)
ucell = proto.get('ucell')
masses = pot.masses(elements)

atomtypes = []
for i in xrange(len(elements)):
    atomtypes.append(iprp.AtomType(elements[i],masses[i]))
pbc = [True, True, True]
 
#Read first neighbor divacancy information    
nn_params = ptd.get('1nn','parameters')
if nn_params == None:
    raise ValueError('1nn divacancy information not found!')

if size%2 ==0:        
    i = size/2
    sys_size = np.array([[-i,i],[-i,i],[-i,i]], dtype=np.int)
    
sys0 = iprp.lammps.create_sys(lammps_exe, atomtypes, pbc, alat=alat, ucell=ucell, size=sys_size)

#Construct divacancy system, dvsys 
pos1 = np.array([alat[0]*nn_params[0]['pos'][0],
                 alat[1]*nn_params[0]['pos'][1],
                 alat[2]*nn_params[0]['pos'][2]])

pos2 = np.array([alat[0]*nn_params[1]['pos'][0],
                 alat[1]*nn_params[1]['pos'][1],
                 alat[2]*nn_params[1]['pos'][2]])

psys, dvsys = sys0.pt_defect(ptdtype='v', pos=pos1)
psys, dvsys = dvsys.pt_defect(ptdtype='v', pos=pos2)

#Construct end vacancy systems by adding one atom back in to the divacancy systems
psys, vsys1 = dvsys.pt_defect(ptdtype='i', pos=pos1, atype=1)
psys, vsys2 = dvsys.pt_defect(ptdtype='i', pos=pos2, atype=1)

#Perform NEB calculation            
vm_rcoord, vm_energy = neb_vacancy(vsys1, vsys2, lammps_exe, pot, elements, masses, num_replicas)
matplotlib.rc('lines',lw=2)
matplotlib.rc('font',size=15)
plt.figure(figsize=(10,8))
plt.plot(vm_rcoord,vm_energy,'b-o',label='Vacancy Mobility Barrier')

plt.legend(loc='lower right')
plt.xlabel('Reaction Coordinate',size='x-large')
plt.ylabel('Vacancy mobility barrier energy (eV)',size='x-large')
plt.show()
print 'Vacancy mobility: %.3f eV'%max(vm_energy)    
```
