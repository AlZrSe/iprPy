###1.3 Calculation Parameters

These are user-specified parameters that control how the calculation proceeds, such as number of data points.

- temp_min, temp_max and temp_delta set the temperature range to be explored.

- run_steps is the total number of MD iteration steps to run at each temperature. Increasing the run time adds to the number of data points being averaged and can help improve the statistical analysis.

- relax_steps is the number of MD iteration steps at the beginning of the run to ignore.  This allows for the data during the initial relaxation period to be excluded. 

- dump_every sets how often LAMMPS dump files of the atoms and calculated parameters are created.  By default, setting dump_every = run_steps means only two files are created: one for the initial configuration and one for the final. This helps in the analysis associated with determining if the crystal structure has transformed. 

- thermo_steps sets how often the system-wide measurements are made.  Decreasing this number adds to the number of data points being averaged, but doesn't add to the total run time.

- size sets the dimensions of the system that is created.  Increasing the system size decreases the deviation noise for the values measured, but also greatly increases the calculation time. 

- random_seed is an integer that is used by LAMMPS in generating random numbers for assigning velocities.  If set to None, a random number is assigned to the random_seed.

##3. LAMMPS Script Generation Function(s)

This code generates the underlying LAMMPS script(s) for performing the necessary simulations.  

- temp_alat_script(). Uses iprp.lammps.sys_gen_script() to create a system based on the supplied crystal prototype unit cell (ucell), lattice parameters (alat), and system size (size). Then, the atomic positions and system dimensions are relaxed to zero pressure according to the rest of the input parameters.  If temperature = 0, the relaxation uses NPH integration with langevin energy dampening.  Otherwise, the relaxation uses NPT integration.  

```python
#Create lammps script that allows for nph (at temperature = 0K) or npt (at temperature > 0K) relaxation
def temp_alat_script(pair_info, units, atom_style, masses, ucell, alat, 
                     size = np.array([5,5,5], dtype=np.int), temperature = 0, 
                     run_steps = 100000, dump_every = 100000, thermo_steps = 1000,
                     random_seed = None):
    
    boxsize = np.array([[0,size[0]], [0,size[1]], [0,size[2]]], dtype=np.int)
    sys_gen = iprp.lammps.sys_gen_script(alat=alat, ucell=ucell, size=boxsize)
    
    mass = ''
    for i in xrange(0,len(masses)):
        mass += 'mass %i %f\n'%(i+1, masses[i])
    
    if random_seed == None:
        random_seed = random.randrange(1,1000000)
        
    nl = '\n'
    script = nl.join(['boundary p p p',
                      'units '+ units,
                      'atom_style ' + atom_style,
                      '',
                      sys_gen,
                      '',
                      mass,
                      pair_info,
                      '',
                      'compute pe all pe/atom',
                      'compute ke all ke/atom',
                      'compute stress all stress/atom NULL',
                      'compute ackland all ackland/atom',
                      'compute rdf all rdf 100',
                      'compute voro all voronoi/atom surface all',
                      '',
                      'thermo %d' % (thermo_steps),
                      'thermo_style custom step temp pe ke etotal lx ly lz pxx pyy pzz pyz pxz pxy',
                      'timestep 0.001',
                      ''])
    
    if temperature == 0.0:
        script = nl.join([script,
                          'fix nph all nph aniso 0.0 0.0 1.0',
                          'fix langevin all langevin 0.0 0.0 1.0 %d' % (random_seed)])
    else:
        script = nl.join([script,
                          'velocity all create %lf %d' % (2*temperature, random_seed),
                          'fix npt all npt temp %lf %lf 0.1 aniso 0.0 0.0 1.0' % (temperature, temperature)])

    script = nl.join([script,
                      'dump dump_chkpt all custom %d dump.* id type mass xu yu zu &' % dump_every,
                      'vx vy vz &',
                      'c_pe c_ke &',
                      'c_ackland &',
                      'c_voro[1] c_voro[2] c_voro[3] &',
                      'c_stress[1] c_stress[2] c_stress[3] c_stress[4] c_stress[5] c_stress[6]',
                      '',
                      'fix print_rdf all ave/time %d 1 %d c_rdf file rdf.dat mode vector' % (dump_every, dump_every),
                      ''
                      'run %d' % run_steps])
    return script
```

##4. Python Calculation Function(s)

These functions are associated with setting up the atomic systems, running LAMMPS, extracting the simulation data, and refining the results.

- alat_relax() sets up the relaxation simulation, runs it, and extracts the results.  The function returns the average temperature and lattice parameter values measured, along with standard deviation errors for both.

```python
def alat_relax(lammps_exe, pot, elements, masses, proto, alat_init, 
               size = np.array([5,5,5], dtype=np.int), temperature = 0, 
               run_steps = 100000, dump_every = 100000, thermo_steps = 100, relax_steps = 1000,
               random_seed = None):
    
    #Get parameters from pot
    units = pot.get('units')
    pair_info = pot.coeff(elements)
    atom_style = pot.get('atom_style') 
    
    #Get parameters from proto
    ucell = proto.get('ucell')
    
        
    #Perform the simulation
    with open('temp_alat.in','w') as f:
        f.write(temp_alat_script(pair_info, units, atom_style, masses, ucell, alat0, 
                                 size=size, temperature=temperature, thermo_steps=thermo_steps,
                                 run_steps=run_steps, dump_every=dump_every, random_seed=random_seed))

    data = iprp.lammps.extract(subprocess.check_output(lammps_exe+' < temp_alat.in',shell=True))
    
    times = []
    tempers = []
    a_params = []
    b_params = []
    c_params = []

    for i in xrange(1, len(data)):
        times.append(int(data[i][0]))
        tempers.append(float(data[i][1]))
        a_params.append(float(data[i][5]) / size[0])
        b_params.append(float(data[i][6]) / size[1])
        c_params.append(float(data[i][7]) / size[2])
        
    start = times.index(relax_steps)
    
    #Compute temperature mean and deviation
    temperature = np.array([np.mean(tempers[start:]), 2*np.std(tempers[start:])])
    
    #Compute lattice parameter mean and deviation
    alat = np.array([[np.mean(a_params[start:]), 2*np.std(a_params[start:])],
                     [np.mean(b_params[start:]), 2*np.std(b_params[start:])],
                     [np.mean(c_params[start:]), 2*np.std(c_params[start:])]])
                       
    return temperature, alat 
       
```

###5.1 Run Calculations

This section performs the calculations.

```python
#Read potential information
pot = iprp.lammps.Potential(pot_file, pot_dir)

#Read crystal prototype information
proto = iprp.models.CrystalPrototype(proto_file)

#Build masses list 
if proto.get('nsites') == len(elements):
    masses = []
    for el in elements:
        masses.append(iprp.get_mass(el))
else:
    raise ValueError('Number of elements listed does not match number of prototype sites')

#Setup alat0 values    
if isinstance(alat0, (int, long, float)):
    alat0 = alat0 * proto.get('lat_mult')
elif isinstance(alat0, (list, tuple)) and len(alat0) == 3:
    alat0 = np.array(alat0)
elif not (isinstance(alat0, np.ndarray) and len(alat0) == 3):
    raise TypeError('Unknown format for alat0')
    
#initialize the data arrays for plotting 
temps = []
temp_errs = []
avals = []
aval_errs = []
bvals = []
bval_errs = []
cvals = []
cval_errs = []
vols = []
vol_errs = []

#Loop over the temperature range
for set_temp in xrange(temp_min, temp_max + temp_delta, temp_delta):
    try:
        os.chdir(os.path.join(working_dir, str(set_temp)))
    except:
        os.mkdir(os.path.join(working_dir, str(set_temp)))
        os.chdir(os.path.join(working_dir, str(set_temp)))
        
    #Call the calculation function
    temperature, alat = alat_relax(lammps_exe, pot, elements, masses, proto, alat0, 
                                   size = size, temperature = set_temp, run_steps = run_steps, 
                                   dump_every = dump_every, thermo_steps = thermo_steps, 
                                   relax_steps = relax_steps, random_seed = random_seed)

    #Calculate the atomic volume and associated error using the lattice parameters
    vol = (alat[0][0] * alat[0][0] * alat[0][0]) / len(proto.get('ucell'))
    vol_err = abs(vol) * ((alat[0][1]/alat[0][0])**2 + (alat[1][1]/alat[1][0])**2 + (alat[2][1]/alat[2][0])**2 )**0.5
 
    #Add the calculated data to arrays for plotting
    temps.append(temperature[0])
    temp_errs.append(temperature[1])
    avals.append(alat[0][0])
    aval_errs.append(alat[0][1])
    bvals.append(alat[1][0])
    bval_errs.append(alat[1][1]) 
    cvals.append(alat[2][0])
    cval_errs.append(alat[2][1])
    vols.append(vol)
    vol_errs.append(vol_err)
    
    #Display and save data to files
    output = '\n'.join(['Temp = %f +- %f K' % (temperature[0], temperature[1]),
                        'alat = %f +- %f A' % (alat[0][0], alat[0][1]),
                        'blat = %f +- %f A' % (alat[1][0], alat[1][1]),
                        'clat = %f +- %f A' % (alat[2][0], alat[2][1]),
                        'A.V. = %f +- %f A^3 / atom' % (vol, vol_err)])
    with open('output.txt', 'w') as f:
        f.write(output)
    print output
    print
os.chdir(working_dir)
```
