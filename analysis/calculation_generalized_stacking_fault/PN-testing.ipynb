```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('font',family='Times New Roman')
mpl.rc('lines',linewidth=3)
```

```python
rootdir = "C:\\Users\\lmh1\\Documents\\Python-packages\\iprPy\\test\\dislocation_Peierls_Nabarro"
```

```python
system_model = os.path.join(rootdir, 'Ackland_Fe', 'bcc1', 'af3f9bd2-826a-45ca-b09f-3fb19714814b.xml')
gamma_model =  os.path.join(rootdir, 'Ackland_Fe', 'bcc1', '97c1866c-c34e-40a8-bee1-b7bc0ccd520e.xml')

system = am.load('system_model', system_model, key='relaxed-atomic-system')[0]
gamma = am.defect.GammaSurface(model=gamma_model, box=system.box)
gamma.E_gsf_surface_plot()
plt.show()
```
